```python
def cari_bprice(skun):
    import re
    namasku = skun
    for i in range(len(bottom)):
        sk = bottom['sku'][i]
        x = re.search(sk, namasku)
        if x:
            return(bottom['bottom_price'][i])
```

```python
list_inkosisten = ["mirip","bukan","seperti","versi lama","new version","versi baru","model lama","model baru","murah","mahal","paket","package","bundling","bundle"]
```

```python
# tandai listing inconsistent dengan membuat sebuah fungsi
def cari_inkosisten(kata):
    import re
    katas = kata
    for i in range(len(list_inkosisten)):
        sk = list_inkosisten[i]
        x = re.search(sk, katas)
        if x:
            return("inconsistent")
        else:
            return(np.nan)
```

```python
#Perulangan untuk mencari judul yang inkosisten
for i in range(len(Gabungan3)):
    judul = Gabungan3["Product Title"][i]
    inkosis_judul = cari_inkosisten(judul)
    #print(inkosis_judul)
    Gabungan3['Inconsistent Product Title'][i] = inkosis_judul
```
