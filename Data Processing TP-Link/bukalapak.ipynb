```python
import pandas
import numpy as np
```

```python
directory = 'bukalapak/'
filename = os.listdir(directory)
file_directory = os.path.join(directory, filename[0])
df = pandas.read_csv(file_directory)
```

```python
my_list = ['tplink', 'tp link', 'tp_link', 'tp-link', 'tplin/k', 'mercusys']
dsa_str = '|'.join(my_list)
df = df[df['name'].astype(str).str.contains(dsa_str, case = False).any(level=0)]
```

```python
df['name'] = df['name'].str.replace('braket','')
df['name'] = df['name'].str.replace('bracket','')
df['name'] = df['name'].str.replace('tatakan','')
df['name'] = df['name'].str.replace('dudukan','')
df['name'] = df['name'].str.replace('hanger','')
df['name'] = df['name'].str.replace('holder','')
```

```python
df = df[['name','price','url','available','seller_name']]
```

```python
df.columns = ['title', 'Selling/After Discount Price', 'url', 'In Stock', 'Seller Name']
```

```python
df.insert(0,'SKU Name', np.nan)
df.insert(2, 'Bottom Price', np.nan)
df.insert(3, 'Original Price', df[['Selling/After Discount Price']])
df.insert(5,'% Discrepancy Bottom Price vs Selling Price', np.nan)
df.insert(6, 'Merchant Name', 'bukalapak')
df.insert(9, 'Total Sold', np.nan)
```

```python
df["In Stock"].replace({True: "In Stock", False: "Out of Stock"}, inplace=True)
```

```python
df = df.drop_duplicates(subset=['url'], keep='first')
```

```python
df.reset_index(drop=True, inplace=True)
```
